### Frame the Problem

The purpose of this project is to be able to predict the final price of each home in Ames, Iowa, given a set of explantory variables.
The data that has been provided has 79 explanatory variables that describe almost every aspect of the house. These variables can be used to predict the final price of each home.

This project, for example, may be used to help property investors make better and more optimal decisions for their investment. Or for people to make a more accurate prediction of the selling price of a house. 

Performance will be evaluated on Root-Mean-Squared-Error (RMSE) between the logarithm of the predicted value and the logarithm of the observed sales price. (Taking logs means that errors in predicting expensive houses and cheap houses will affect the result equally.) Reducing the RMSE results in a more accurate prediciton price of the house.

## Explore the Data and Gain Insight

Exploratory Analysis has been done so we will now start visualising the data and look for interesting relationships between the sales price and the explanatory variables.

In this part of the project we will create visualisations of the data to create meaningful insights. 

First we will explore our response variable

```python
%matplotlib inline

def explore(feature, 
            workaround=False # circumvent error in matplotlib
           ):
    gridkw = dict(height_ratios=[5, 1])
    fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw=gridkw, sharex=True)
    sns.distplot(feature, ax=ax1, kde=False) #array, top subplot
    sns.boxplot(feature, ax=ax2, width=.4) #bottom subplot
    ax1.set_xlabel('') 
    if workaround:
        print(feature.describe())
    else:
        ax1.text(1.05, 0.95, 
             feature.describe(), 
             transform=ax1.transAxes, fontsize=14,
            verticalalignment='top')
    fig.subplots_adjust(top=0.9)
    fig.suptitle(feature.name, fontsize=16)
```

```python
explore(housingData.SalePrice, workaround = True)
```

The sale price of houses in the Ames, Iowa area have a mean of $\$180,923$. The sales prices are slightly right skewed meaning that the tale tends towards larger sale prices. $50\%$  of the sales prices are between $\$130,000$ and $\$214,000$.

#### Exploring the Correlations 

```python
# Inspect how the attributes are correlated to the sale price
corr_matrix = housingData.corr()
corr_matrix["SalePrice"].sort_values(ascending=False)
```

#### Visualisation of the correlations  

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

The correlation plot shows that the sale price has a strong correlation with many of the variables. However, a lot of these have strong correlations with one another so this could be an example of multi-collinearity and these extra variables are not needed to help explain the sales price (such as Area of Garage, and number of cars in the Garage) 

#### Visualise important correlations graphically

```python
sns.pairplot(housingData[["SalePrice", "TotRmsAbvGrd", "GrLivArea", "LotArea", 
                           "OverallQual", "GarageArea", "Neighborhood"]],
             hue="Neighborhood",
             diag_kind='hist',
            plot_kws= { "alpha":0.4})
```

From the plots above we can see that there are strong trends between sales price, and overall quality and the above ground living area. The neighbourhood was used to partition the plots even more, however, there appears to be no strong or obvious trend between sales price and the neighbourhood.

```python
f, axes = plt.subplots(1, 2)

sns.countplot(x = "MoSold", data = housingData, ax=axes[0])
sns.countplot(x = "YrSold", data = housingData, ax=axes[1])
```

The plots above show the distrubution of the sales through the year and the month. THe most popular month for selling a house appears to be the summer time (May - Jun). 

There appers to be a very steady sale of houses between the years 2006 and 2009. There is a large drop in the number of houses in the year 2010. This could be due to the data bein collected halfway through they year hence not having the entire year for 2010. 

```python
sns.catplot(x = "YrSold", y = "SalePrice", aspect = 3.0, kind = "point", data = housingData)
```

The plots above show how the sales price changes over the month it was sold and the year it was sold. We can see that, on average, the houses sold in 2007 are more expensive than other years. 

Houses sold in the second half of the year are roughly $\$15,000$ more expensive than that of houses sold in the first half of the year. 

```python
sns.countplot(x = "YearBuilt", data = housingData)

sns.catplot(x = "YearBuilt", y = "SalePrice", aspect = 2.0, kind = "point", data = housingData)
```

From the two plots above, showing the number of houses built in each year and the sales price of houses over the year built, we can see that both of these trends increase. Generally, there rate at which houses have been built are increasing. Also, the price at which houses are selling for are increasing as the year they are built increases (with the odd few years with large sale prices). 

The plot below shows the general trend of the sale price for the house increasing as the year it was built increases. This plot also shows the density and where most of the values occur.
